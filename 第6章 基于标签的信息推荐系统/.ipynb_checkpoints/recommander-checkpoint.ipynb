```python
import html2text as html2text
from bs4 import BeautifulSoup
import re
import csv
```

```python
with open(r"G:\proprocess\第6章 基于标签的信息推荐系统\news.csv",encoding='utf-8') as file:
    reader = csv.reader(file)
    news_list = list(reader)
    
```

```python
news_list[1][15][:1000]
```

```python
#移除特殊符号
def remove_punc(line_sentence):
    multi_version = re.compile("-\{.*?(zh-hans|zh-cn):([^;]*?)(;.*?)?\}-")
    punctuation = re.compile("[-~!@#$%^&*()_+`=\[\]\\\\\n\\t{\}\\\r|;':,./<>?·！@#￥%……&*（）——+【】▋、；‘：“”，。、《》？「『」』]")
    line = multi_version.sub(r"\2", line_sentence)
    line = punctuation.sub(' ', line_sentence)
    return line
```

```python
new_list = []
for row in news_list[:2]:
    cleantext = BeautifulSoup(row[15], "lxml").text
    new_list.append(remove_punc(cleantext))
new_list
```
